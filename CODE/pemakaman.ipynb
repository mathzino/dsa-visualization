```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_pemakaman  = pd.read_csv("../data/Pemakaman.csv")
data_pemakaman.fillna(0, inplace=True)

data_pemakaman.describe()
```

```python
data_pemakaman['Tanggal'].str.strip()
data_pemakaman['Tanggal'] = pd.to_datetime(data_pemakaman['Tanggal'], format='%d/%M/%Y',exact=False).dt.date
data_pemakaman.head(100)
```

```python
column_data_pemakaman=[ "Jumlah Laporan", "Akumulasi", "Harian Pemakaman Umum", "Total Pemakaman Umum"]
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(10, 10))
for i, column in enumerate(column_data_pemakaman):
    sns.lineplot(data_pemakaman[column], ax = ax[i//2, i%2]) 
    ax[i//2, i%2].set_title("line of {}".format(column),fontsize=11)

# plt.tight_layout() 
plt.show()
```
