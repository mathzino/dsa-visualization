```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_jakarta_indonesia_covid  = pd.read_csv("../data/Data Indonesia dan Jakarta.csv")
data_jakarta_indonesia_covid.fillna(0, inplace=True)
data_jakarta_indonesia_covid
```

```python
data_jakarta_indonesia_covid['Tanggal'].str.strip()
data_jakarta_indonesia_covid['Tanggal'] = pd.to_datetime(data_jakarta_indonesia_covid['Tanggal'], format='%d/%M/%Y',exact=False).dt.date
for column in data_jakarta_indonesia_covid.columns:
    for index, row in data_jakarta_indonesia_covid.iterrows():
        value = row[column]
        if isinstance(value, str):
            cleaned_value = value.replace(",", "")
            converted_value = int(float(cleaned_value))
            data_jakarta_indonesia_covid.at[index, column] = converted_value
    
```

```python

column_data_Indonesia=[ 'Positif (Indonesia)', 'Dirawat (Indonesia)',
       'Sembuh (Indonesia)', 'Meninggal (Indonesia)', 'Meninggal (Jakarta)',
       'Sembuh (Jakarta)', 'Self-Isolation (Jakarta)', 'Dirawat (Jakarta)',
       'Positif (Jakarta)', 'Positif Aktif (Jakarta)',
       'Positif Harian (Indonesia)', 'Dirawat Harian (Indonesia)',
       'Sembuh Harian (Indonesia)', 'Meninggal Harian (Indonesia)',
       'Meninggal Harian (Jakarta)', 'Sembuh Harian (Jakarta)',
       'Self-Isolation Harian (Jakarta)', 'Dirawat Harian (Jakarta)',
       'Positif Harian (Jakarta)', 'Positif Aktif Harian (Jakarta)']

fig, ax = plt.subplots(ncols=2, nrows=10, figsize=(10, 35))
for i, column in enumerate(column_data_Indonesia):
    sns.lineplot(data_jakarta_indonesia_covid[column] , ax=ax[i//2, i%2] )
    ax[i//2, i%2].set_title("line of {}".format(column),fontsize=11)

# plt.tight_layout() 
plt.show()
```
